```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.indexmundi.com/commodities/?commodity=gold&months=360&currency=myr"
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[1]
```

```python
soup.find("table", class_ = "tblData")
```

```python
table = soup.find_all("table")[1]
print(table)
```

```python
#get title or th tag
gold_titles = soup.find_all("th")
```

```python
gold_titles
```

```python
gold_titles = [title.text for title in gold_titles]
```

```python
gold_titles
```

```python
import pandas as pd

#create title for data frame
df = pd.DataFrame(columns = gold_titles)
```

```python
df
```

```python
#to get data in table
table.find_all('tr')
```

```python
column_data = table.find_all("tr")
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    each_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = each_row_data

df
```

```python
df.info()
```

```python
#to change datatype from object to float, but change to string first
import pandas as pd

df['Price'] = pd.Series(df['Price'], dtype="string")
```

```python
df.dtypes
```

```python
#To remove comma
df['Price']=df['Price'].str.replace(',','')
df
```

```python
#To change to float
df['Price'] = df['Price'].astype(float)
```

```python
df
```

```python
df.dtypes
```

```python
df.describe()
```

```python
#To change 'Month' datatype to datetime, but change to string first
df['Month'] = pd.Series(df['Month'], dtype="string")
df
```

```python
df.dtypes
```

```python
from datetime import datetime
df['Month'] = df['Month'].apply(lambda x: datetime.strptime(x, "%b %Y"))
df
```

```python
#To display only Month and Year 
df['Month'] = df['Month'].dt.to_period('M')
df
```

```python
#To change 'Change' datatype to float but change to string first
df['Change'] = pd.Series(df['Change'], dtype="string")
```

```python
df['Change']=df['Change'].str.replace('%','')
df["Change"] = df["Change"].str.replace('','-').str.replace(r'^-$', '0', regex=True).astype(float)
```

```python
df
```

```python
df.dtypes
```

```python
df['Change'] = df['Change'] / 100;
df
```

```python
#troy ounce to gram
gram = 31.1035

#Price per gram
df['Price'] = df['Price']/gram
df
```

```python
df.dtypes
```

```python
df.astype
```

```python
#Reset the index and drop index
df = df.reset_index(drop=True)
```
